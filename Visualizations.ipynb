```python
#importing libraries 
import numpy as np
from hockey_rink import NHLRink, RinkImage
from hockey_rink import Rink
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Linhac_df_keyed_20_games.csv')
```

## All these visuals are of the base dataset.

###  Plot of all shots taken 

```python
outcome = {'successful': (0, 1, 0, 1), 'failed': (1,0,0,1)}

df_shot = df[df['eventname'] == 'shot']

rink = Rink()
fig, ax = plt.subplots(figsize=(14, 10))
rink.scatter("xadjcoord", "yadjcoord", facecolor=df_shot.outcome.map(outcome), s=30, edgecolor="white", data=df_shot, ax=ax)
```

###  Plot of all goals taken 

```python
outcome = {'successful': (0, 1, 0, 1), 'failed': (1,0,0,1)}
df_goal = df[df['eventname'] == 'goal']

rink = Rink()
fig, ax = plt.subplots(figsize=(14, 10))
rink.scatter("xadjcoord", "yadjcoord", facecolor=df_goal.outcome.map(outcome), s=50, edgecolor="white", data=df_goal, ax=ax)
```

###  All shots  taken

```python
#chart of all shots
shots = (
    df.assign(
        is_shot=df['eventname'] == "shot",
        x=np.abs(df['xadjcoord']),
        y=df['yadjcoord'] * np.sign(df['xadjcoord']),
    )
)


fig, axs = plt.subplots(1, 3, figsize=(18, 8))

rink = NHLRink(rotation=270, net={"visible": False})

rink.contourf(
    "x", "y", "is_shot", data=shots, 
    nbins=8, levels=30, plot_range="ozone", cmap="bwr",
    ax=axs[0], draw_kw={"display_range": "ozone"},
)

rink.heatmap(
    "x", "y", "is_shot", data=shots, 
    binsize=5, fill_value=0, plot_xlim=(25, 89), cmap="magma", vmax=0.25,
    ax=axs[1], draw_kw={"display_range": "ozone"},
)

rink.hexbin(
    "x", "y", "is_shot", data=shots,
    gridsize=(14, 8), plot_range="ozone", alpha=0.85, vmax=0.25,
    ax=axs[2], draw_kw={"display_range": "ozone"},
)
```

###  All goals made

```python
#chart of all goals
goals = (
    df.assign(
        is_goal=df['eventname'] == "goal",
        x=np.abs(df['xadjcoord']),
        y=df['yadjcoord'] * np.sign(df['xadjcoord']),
    )
)


fig, axs = plt.subplots(1, 3, figsize=(18, 8))

rink = NHLRink(rotation=270, net={"visible": False})

rink.contourf(
    "x", "y", "is_goal", data=goals, 
    nbins=8, levels=30, plot_range="ozone", cmap="bwr",
    ax=axs[0], draw_kw={"display_range": "ozone"},
)

rink.heatmap(
    "x", "y", "is_goal", data=goals, 
    binsize=5, fill_value=0, plot_xlim=(25, 89), cmap="magma", vmax=0.25,
    ax=axs[1], draw_kw={"display_range": "ozone"},
)

rink.hexbin(
    "x", "y", "is_goal", data=goals,
    gridsize=(14, 8), plot_range="ozone", alpha=0.85, vmax=0.25,
    ax=axs[2], draw_kw={"display_range": "ozone"},
)
```

###  All assists

```python
#chart of all assists
assists = (
    df.assign(
        is_assist=df['eventname'] == "assist",
        x=np.abs(df['xadjcoord']),
        y=df['yadjcoord'] * np.sign(df['xadjcoord']),
    )
)


fig, axs = plt.subplots(1, 3, figsize=(18, 8))

rink = NHLRink(rotation=270, net={"visible": False})

rink.contourf(
    "x", "y", "is_assist", data=assists, 
    nbins=8, levels=30, plot_range="ozone", cmap="bwr",
    ax=axs[0], draw_kw={"display_range": "ozone"},
)

rink.heatmap(
    "x", "y", "is_assist", data=assists, 
    binsize=5, fill_value=0, plot_xlim=(25, 89), cmap="magma", vmax=0.25,
    ax=axs[1], draw_kw={"display_range": "ozone"},
)

rink.hexbin(
    "x", "y", "is_assist", data=assists,
    gridsize=(14, 8), plot_range="ozone", alpha=0.85, vmax=0.25,
    ax=axs[2], draw_kw={"display_range": "ozone"},
)
```
